```python
df_high_rate = df[df["competing1"] > 0.9]
nb_of_default = df_high_rate[df_high_rate["default"] == 1]
print("Number of defaults when Team 1 rate > 90% :", len(nb_of_default), "out of", len(df_high_rate))
print("Proportion of defaults when Team 1 rate > 90% :", len(nb_of_default) / len(df_high_rate) * 100)
print("return on high rate loans :", nb_of_default["profit"].mean())
```

```python
print("avg interest rate for Team 1 :", df["competing1"].mean() * 100)
print("avg interest rate for Team 3 :", df["competing3"].mean() * 100)
print("Our avg interest rate:", df["own"].mean() * 100)
```

```python
# 1. Trouver la valeur maximale pour chaque ligne
row_max = df[["own", "competing1", "competing3"]].max(axis=1)

# 2. Créer les deux conditions de filtrage
# Condition 1: competing1 est le max de la ligne
is_comp1_max = df["competing1"] == row_max
is_comp3_max = df["competing3"] == row_max

# Condition 2: Le client a fait défaut
is_default = df["default"] == True

# 3. Appliquer les deux filtres avec '&' et calculer la moyenne
avg_interest_rate_default_compet1 = df[is_comp1_max & is_default]["competing1"].mean()
avg_interest_rate_default_compet3 = df[is_comp3_max & is_default]["competing1"].mean()

interest_on_default = df[df["profit"] == -10000]["own"].mean()

print(
    "avg interest rate on default for Team 1 :", avg_interest_rate_default_compet1 * 100
)
print(
    "avg interest rate on default for Team 3 :", avg_interest_rate_default_compet3 * 100
)
print("avg interest rate on default us :", interest_on_default * 100)
```

```python
print(
    "diff between avg rate and avg rate on default Team 1 :",
    (avg_interest_rate_default_compet1 - df["competing1"].mean()) * 100,
)
print(
    "diff between avg rate and avg rate on default Team 3 :",
    (avg_interest_rate_default_compet3 - df["competing3"].mean()) * 100,
)
print(
    "diff between avg rate and avg rate on default us :",
    (interest_on_default - df["own"].mean()) * 100,
)
```

```python
df = pd.read_csv('data/Round1_Diagnostic_21.csv')
df.head()
```

```python
df
```

```python
def grant_to(row):
    if row["borrowertype"] == 1:
        if min(row["own"], row["competing1"] - 0.02, row["competing3"]) == row["competing1"] - 0.02:
            return 1
        if min(row["own"], row["competing1"] - 0.02, row["competing3"]) == row["competing3"]:
            return 3
        return 2
    if row["borrowertype"] == 2:
        if min(row["own"] - 0.02, row["competing1"], row["competing3"]) == row["own"] - 0.02:
            return 2
        if min(row["own"] - 0.02, row["competing1"], row["competing3"]) == row["competing3"]:
            return 3
        return 1
    if row["borrowertype"] == 3:
        if min(row["own"], row["competing1"], row["competing3"] - 0.02) == row["own"]:
            return 2
        if min(row["own"], row["competing1"], row["competing3"] - 0.02) == row["competing1"]:
            return 1
        return 3
```

```python
def profit1(row):
    if row["granted_to"] == 1:
        if row["default"] == 1:
            return -10000
        else:
            return 10000 * row["competing1"]
    else:
        return None
```

```python
def profit3(row):
    if row["granted_to"] == 3:
        if row["default"] == 1:
            return -10000
        else:
            return 10000 * row["competing3"]
    else:
        return None
```

```python
def profit2(row):
    if row["granted_to"] == 2:
        if row["default"] == 1:
            return -10000
        else:
            return 10000 * row["own"]
    else:
        return None
```

```python
df["granted_to"] = df.apply(grant_to, axis=1)
```

```python
df["profit1"] = df.apply(profit1, axis=1)
df["profit3"] = df.apply(profit3, axis=1)
```

```python
df
```

```python
print("total profit Team 1:", df["profit1"].sum())
print("total profit Team 2:", df["profit"].sum())
print("total profit Team 3:", df["profit3"].sum())
```
